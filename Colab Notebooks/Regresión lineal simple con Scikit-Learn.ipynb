```
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```
dataPATH = '/content/salarios_32849f3e-7f53-4139-b334-b8de8beb144a.csv'

dataset = pd.read_csv(dataPATH)
```

```
dataset
```

```
x = dataset.iloc[: ,:-1].values
y = dataset.iloc[:,1].values
print(y)
print(x)
```

la siguiente funcion divide los datos del test y del entrenamiento automaticamente

```
X_train , X_test , Y_train , Y_test = train_test_split(x , y ,test_size = 0.2 , random_state = 0)
```

#regresion lineal

```
model = LinearRegression()

model.fit(X_train , Y_train)
```

```
plt.scatter(X_train , Y_train , color = 'blue')
plt.plot(X_train , model.predict(X_train) , color = 'black')
```

```
plt.scatter(X_test , Y_test , color = 'blue')
plt.plot(X_train , model.predict(X_train) , color = 'black')
```

```
model.score(X_test , Y_test) #nos dice que tan bueno fue el modelo con los datos de prueba que le dimos (X_test) 
```
