```
import pandas as pd
from sklearn.cluster import MiniBatchKMeans # es casi lo mismo que k_means pero consume muchos menos recursos , funciona muy bien para datasets peque√±os
```

```
if __name__ == "__main__":

  dataset_candy_PATH = '/content/drive/My Drive/curso_sklearn/datasets/candy.csv'
  dataset_candy = pd.read_csv(dataset_candy_PATH)

  print(dataset_candy.shape)
  print(dataset_candy.head(10))

  X = dataset_candy.drop('competitorname' , axis=1)

  kmeans = MiniBatchKMeans(n_clusters= 4 , # lo agrupamos en 4 grupos
                           batch_size = 8)
  kmeans.fit(X)

  print('total de centros: ' , len(kmeans.cluster_centers_))
  print('='*64)
  dataset_candy['group'] = kmeans.predict(X)

  print(dataset_candy.head(10))
```
