```
import pandas as pd

from sklearn.model_selection import RandomizedSearchCV
from sklearn.ensemble import RandomForestRegressor
```

```
if __name__ == "__main__":

  dataset_PATH = '/content/drive/My Drive/curso_sklearn/datasets/felicidad.csv'
  dataset_felicidad = pd.read_csv(dataset_PATH)

  X = dataset_felicidad.drop(['country' ,'rank', 'score'] , axis = 1)
  Y = dataset_felicidad[['score']]

  print(Y)  
  
```

```
  reg = RandomForestRegressor()

  parametros = { # ponemos los parametros que queremos probar en un diccionario   
    'n_estimators' : range(4,16) ,
    'criterion' : ['mse' , 'mae'] ,
    'max_depth' : range(2,11)
  }

  rand_est = RandomizedSearchCV(reg , parametros , cv = 3  , scoring = 'neg_mean_absolute_error',
                               n_iter = 10 # maximo numero de configuraciones (solo toma 10 convinaciones de los parametros)
                               )

  rand_est.fit(X , Y)
  print('='*64)
  print(rand_est.best_estimator_)
  print('='*64)
  print(rand_est.best_params_)
```

```
  print(rand_est.predict(X.loc[[0]]))
```
