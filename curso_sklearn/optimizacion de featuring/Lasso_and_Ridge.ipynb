# Lasso and Ridge

```
import pandas as pd 
import sklearn 
import matplotlib.pyplot as plt 

from sklearn.linear_model import LinearRegression
from sklearn.linear_model import Lasso
from sklearn.linear_model import Ridge

from sklearn.model_selection import  train_test_split
from sklearn.metrics import mean_squared_error
```

```
if __name__ == "__main__":

  datasetPATH = '/content/drive/My Drive/curso_sklearn/datasets/felicidad.csv'

  dataset = pd.read_csv(datasetPATH)
  print(dataset.describe)


  x = dataset[['gdp' , 'family' ,'lifexp' , 'freedom' , 'corruption' , 'generosity' , 'dystopia']]
  y = dataset[['score']]

  print(x.shape , y.shape)
```

```
 X_train , X_tests , Y_train , Y_tests = train_test_split(x,y , test_size = 0.25)
```

**MODELO DE REGRESION LINEAL SIMPLE**

```
  modelLineal = LinearRegression().fit(X_train , Y_train)

  y_predic_linear = modelLineal.predict(X_tests)

  
```

**MODELO LASSO**

```
  modelLasso = Lasso(alpha=0.02).fit(X_train , Y_train)# el alpha es el parametro que podemos ajustar
                                                      #que tan duro penalizamos los features menos importantes
  y_predic_lasso = modelLasso.predict(X_tests)
```

**MODELO RIDGE**

```
  modelRidge = Ridge(alpha=0.1).fit(X_train , Y_train)# el alpha es el parametro que podemos ajustar
                                                      #que tan duro penalizamos los features menos importantes
  y_predic_Ridge = modelRidge.predict(X_tests)
```

comparamos el error medio cuaratico para evaluar las perdidas del modelo 

```
  linear_loss = mean_squared_error(Y_tests , y_predic_linear)

  print('linear loss: ' , linear_loss )


  lasso_loss = mean_squared_error(Y_tests , y_predic_lasso)

  print('lasso loss : ' , lasso_loss )


  linear_ridge = mean_squared_error(Y_tests , y_predic_Ridge)

  print('ridge loss : ' , linear_ridge )
```

```
  print('coef lasso : ' , modelLasso.coef_)
  print('='*60)
  print('coef lineal : ',modelLineal.coef_)
  print('='*60)
  print('coef ridge : ' ,modelRidge.coef_)
  print('='*60)
```
