```
if __name__ == '__main__':

  dataPATH = '/content/drive/My Drive/curso_sklearn/datasets/felicidad_corrupt.csv'

  dataset = pd.read_csv(dataPATH)

  X = dataset.drop(['country' , 'score'] , axis= 1)
  Y = dataset[['score']]

  print(Y)
```

```
  X_train , X_tests , Y_train , Y_tests = train_test_split(X,Y , test_size = 0.3 , random_state = 42)
  
```

ponemos los estimadores en un diccionario para mas comodidad

```
  estimadores = {
      
      'SVR' : SVR(gamma= 'auto', C = 1.0 , epsilon= 0.1),
      'RANSAC' : RANSACRegressor(),
      'HUBER' : HuberRegressor(epsilon= 1.35)
  }
```

```
  for name , estimador in estimadores.items():
    
    estimador.fit(X_train , Y_train)
    prediction = estimador.predict(X_tests)

    print('='*80)
    print(name , "MSE: " , mean_squared_error(Y_tests , prediction))
    
```
