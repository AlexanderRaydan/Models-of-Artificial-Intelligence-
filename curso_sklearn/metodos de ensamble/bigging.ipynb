```
import sklearn
import pandas as pd
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import  BaggingClassifier

from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

```

```

if __name__ == "__main__":

  heart_dataset_PATH = '/content/drive/My Drive/curso_sklearn/datasets/heart.csv'

  heart_dataset = pd.read_csv(heart_dataset_PATH)

  print(heart_dataset['target'].describe())
```

```
  X = heart_dataset.drop(['target'] , axis=1)
  Y = heart_dataset['target']
  print(heart_dataset.shape)
  print(X.shape)
  print(Y.shape)
```

```
  X_train , X_test , Y_train , Y_test = train_test_split(X , Y , test_size = 0.35)
```

```
  Knn_class = KNeighborsClassifier() #un clasificador nolmal para despues mejorar el resultado con un metodo de ensamble
  Knn_class.fit(X_train ,Y_train)
  

  Knn_prediction = Knn_class.predict(X_test)

  print("="*64)
  print(accuracy_score(Knn_prediction , Y_test))
```

ahora lo hacemos con el clasificador por ensamble
(metodo de ensamble)

```
  bag_class = BaggingClassifier(base_estimator= KNeighborsClassifier() , #le mandamos el estimador de base con el cual se va a construir todo
                                n_estimators=50)
  bag_class.fit(X_train , Y_train)
  bag_predict = bag_class.predict(X_test)

  print("="*64)
  print(accuracy_score(bag_predict , Y_test))
```
