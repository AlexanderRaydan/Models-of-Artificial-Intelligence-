```
import sklearn
import pandas as pd
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.ensemble import BaggingClassifier

from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```
if __name__ == "__main__":

  heart_dataset_PATH = '/content/drive/My Drive/curso_sklearn/datasets/heart.csv'

  heart_dataset = pd.read_csv(heart_dataset_PATH)

  print(heart_dataset['target'].describe())
```

```
  X = heart_dataset.drop(['target'] , axis=1)
  Y = heart_dataset['target']
  print(heart_dataset.shape)
  print(X.shape)
  print(Y.shape)
  X_train , X_test , Y_train , Y_test = train_test_split(X , Y , test_size = 0.35)
```

```
  boost = GradientBoostingClassifier(n_estimators= 50) # numeros de arboles para estimar
  boost.fit(X_train , Y_train)
  boost_predict = boost.predict(X_test)
  
  print("="*64)
  print(accuracy_score(boost_predict , Y_test))
```
